Q-1 What is hypothesis testing in statistics ?

SOL - Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. Q-2 What is the null hypothesis, and how does it differ from the alternative hypothesis ?

SOL - Null Hypothesis (𝐻0): The default assumption or status quo in hypothesis testing and also states that there is no effect, no difference, or no relationship between variables as well as It is assumed to be true unless there is strong evidence against it.
Alternative Hypothesis (𝐻1 or 𝐻a): The opposite of the null hypothesis; it claims that there is an effect, difference, or relationship and It is what the researcher aims to prove also If enough evidence is found, we reject

Q-4 What does a P-value represent in hypothesis testing ?

High P-value (𝑝 > 𝛼) → Weak evidence against 𝐻0, so we fail to reject 𝐻0.

Q-1 Write a python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results ?

```python
import numpy as np
from scipy.stats import norm

sample_data = [100, 102, 98, 105, 97, 101, 99, 103, 104, 98]
pop_mean = 100
pop_std = 5
alpha = 0.05

sample_mean = np.mean(sample_data)
n = len(sample_data)

z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject H0: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject H0: No significant difference between the sample and population mean.")
```

Q-4 Perform a two-tailed Z-test using python and visualize the decision region on a plot ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

pop_mean = 100
pop_std = 15
alpha = 0.05

np.random.seed(42)
sample_size = 30
sample_data = np.random.normal(loc=pop_mean, scale=pop_std, size=sample_size)

sample_mean = np.mean(sample_data)

z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

z_critical = norm.ppf(1 - alpha / 2)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-value: ±{z_critical:.4f}")

if abs(z_score) > z_critical:
    print("Reject H0: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject H0: No significant difference between the sample and population mean.")

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

plt.fill_between(x, y, where=(x <= -z_critical), color="red", alpha=0.5, label="Rejection Region")
plt.fill_between(x, y, where=(x >= z_critical), color="red", alpha=0.5)

plt.axvline(-z_critical, color="black", linestyle="dashed", label=f"-Z_critical ({-z_critical:.2f})")
plt.axvline(z_critical, color="black", linestyle="dashed", label=f"Z_critical ({z_critical:.2f})")

plt.axvline(z_score, color="green", linestyle="solid", label=f"Z-score ({z_score:.2f})")

plt.title("Two-Tailed Z-Test Decision Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```

Q-5 Create Python function that calculate and visualizes Type 1 and Type 2 errors during hypothesis testing ?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(alpha=0.05, beta=0.2, mu0=100, mu1=105, sigma=15, sample_size=30):

    standard_error = sigma / np.sqrt(sample_size)

    z_critical = norm.ppf(1 - alpha / 2)
    critical_value = mu0 + z_critical * standard_error

    x = np.linspace(mu0 - 4 * standard_error, mu1 + 4 * standard_error, 1000)

    pdf_h0 = norm.pdf(x, mu0, standard_error)
    pdf_h1 = norm.pdf(x, mu1, standard_error)

    plt.figure(figsize=(10, 5))

    plt.plot(x, pdf_h0, label="H0 (Null Distribution)", color="blue")
    plt.plot(x, pdf_h1, label="H1 (Alternative Distribution)", color="red")

    plt.fill_between(x, pdf_h0, where=(x >= critical_value), color="red", alpha=0.5, label="Type 1 Error (α)")

    plt.fill_between(x, pdf_h1, where=(x < critical_value), color="purple", alpha=0.5, label="Type 2 Error (β)")

    plt.axvline(critical_value, color="black", linestyle="dashed", label=f"Critical Value ({critical_value:.2f})")

    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()

    plt.show()

visualize_type1_type2()
```

Q-6 Write a python programe to perform an independent T-test and interpret the results ?

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)
group1 = np.random.normal(loc=100, scale=15, size=30)
group2 = np.random.normal(loc=105, scale=15, size=30)

t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=True)

alpha = 0.05

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject H0: The two groups have significantly different means.")
else:
    print("Fail to reject H0: No significant difference between the two groups.")
```

Q-9 Write a python function to calculate the confidence interval for a sample mena and explain its significance ?

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):

    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha/2, df=sample_size-1)

    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))


    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return ci_lower, ci_upper

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

ci_lower, ci_upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q-11 Implement a Baysian inference method using bayes'Theoram in python and explain the process ?

```python
def bayesian_inference(prior, likelihood, false_positive_rate, evidence):
    """
    Apply Bayes' Theorem to calculate the posterior probability.

    Args:
    prior: P(H), the prior probability of the hypothesis (e.g., disease)
    likelihood: P(E|H), the likelihood of the evidence given the hypothesis (e.g., accuracy of the test)
    false_positive_rate: P(E|~H), the likelihood of the evidence given the negation of the hypothesis (false positive rate)
    evidence: P(E), the overall probability of the evidence (positive test result)

    Returns:
    posterior: P(H|E), the posterior probability of the hypothesis given the evidence
    """
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.01
likelihood = 0.95
false_positive_rate = 0.05

evidence = (likelihood * prior) + (false_positive_rate * (1 - prior))

posterior = bayesian_inference(prior, likelihood, false_positive_rate, evidence)

print(f"The probability of having the disease given a positive test result (P(H|E)) is: {posterior:.4f}")
```

Q-13 Write a python program to calculate the expected frequencies for a Chi-square test based on observed data ?

```python
import numpy as np

def calculate_expected_frequencies(observed):

    total = np.sum(observed)

    row_sums = np.sum(observed, axis=1, keepdims=True)
    col_sums = np.sum(observed, axis=0, keepdims=True)

    expected = (row_sums * col_sums) / total

    return expected

observed = np.array([[30, 10],
                     [20, 40]])

expected = calculate_expected_frequencies(observed)

print("Observed Frequencies:\n", observed)
print("\nExpected Frequencies:\n", expected)
```

Q-16 Implement an F-test using python to comapare the variances of two random samples ?

```python
import numpy as np
from scipy.stats import f

np.random.seed(42)
sample1 = np.random.normal(0, 1, 100)
sample2 = np.random.normal(0, 2, 100)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F_statistic = var1 / var2 if var1 > var2 else var2 / var1

df1 = len(sample1) - 1
df2 = len(sample2) - 1

p_value = 1 - f.cdf(F_statistic, df1, df2)

print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-statistic: {F_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the Null Hypothesis: The variances are significantly different.")
else:
    print("\nFail to Reject the Null Hypothesis: The variances are not significantly different.")
```

Q-19 Write a python function to check the assumptions (normality, independence, and equal variance) for ANOVA ?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):

    print("Normality Test Results (Shapiro-Wilk):")
    for i, group in enumerate(groups, 1):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i}: Shapiro-Wilk Test Statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"Group {i} does not follow a normal distribution (Reject Null Hypothesis).")
        else:
            print(f"Group {i} follows a normal distribution (Fail to Reject Null Hypothesis).")

    plt.figure(figsize=(12, 8))
    for i, group in enumerate(groups, 1):
        plt.subplot(2, 3, i)
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Group {i} Q-Q Plot")
    plt.tight_layout()
    plt.show()

    print("\nLevene's Test for Homogeneity of Variance:")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test Statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("Reject the null hypothesis: The variances are not equal across groups (Violates assumption).")
    else:
        print("Fail to reject the null hypothesis: The variances are equal across groups (Assumption holds).")

    print("\nIndependence Assumption:")
    print("Independence is assumed based on the random sampling process in the study.")

np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=12, size=30)
group3 = np.random.normal(loc=85, scale=14, size=30)

check_anova_assumptions(group1, group2, group3)
```

Q-20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visulize the results ?

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

n = 60

treatment = np.concatenate([np.repeat('Treatment', n), np.repeat('Control', n)])

age_group = np.concatenate([np.repeat('Young', n), np.repeat('Old', n)])

performance_treatment_young = np.random.normal(loc=70, scale=10, size=n)
performance_treatment_old = np.random.normal(loc=75, scale=10, size=n)
performance_control_young = np.random.normal(loc=80, scale=10, size=n)
performance_control_old = np.random.normal(loc=65, scale=10, size=n)

performance = np.concatenate([performance_treatment_young, performance_treatment_old,
                              performance_control_young, performance_control_old])

data = pd.DataFrame({
    'Treatment': treatment,
    'AgeGroup': age_group,
    'Performance': performance
})

model = ols('Performance ~ C(Treatment) + C(AgeGroup) + C(Treatment):C(AgeGroup)', data=data).fit()
anova_results = anova_lm(model)

print(anova_results)

plt.figure(figsize=(8, 6))
sns.pointplot(x='AgeGroup', y='Performance', hue='Treatment', data=data, dodge=True, markers=["o", "s"])
plt.title('Interaction Plot: Performance by Treatment and Age Group')
plt.xlabel('Age Group')
plt.ylabel('Performance')
plt.grid(True)
plt.show()
```

Q-24 Perform a hypothesis test for opulation variance using a Chi-square distribution and interpret the result ?

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)

sample_size = 30
population_variance = 25
sample_data = np.random.normal(loc=50, scale=np.sqrt(population_variance), size=sample_size)

sample_variance = np.var(sample_data, ddof=1)
print(f"Sample Variance: {sample_variance}")

hypothesized_variance = population_variance

chi_square_statistic = (sample_size - 1) * sample_variance / hypothesized_variance

p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df=sample_size-1), 1 - stats.chi2.cdf(chi_square_statistic, df=sample_size-1))

print(f"Chi-square Statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

Q-25 Write a python script to perform a Z-test for comparing proportions between two datasets or groups ?

```python
import numpy as np
import scipy.stats as stats

x1 = 100
n1 = 200
x2 = 120
n2 = 220

p1_hat = x1 / n1
p2_hat = x2 / n2

p_pool = (x1 + x2) / (n1 + n2)

z_statistic = (p1_hat - p2_hat) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")
```

Q-26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results ?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

f_statistic = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1
df2 = len(data2) - 1

p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

plt.figure(figsize=(10, 6))
plt.hist(data1, bins=20, alpha=0.5, label='Group 1 (Variance = {:.2f})'.format(var1))
plt.hist(data2, bins=20, alpha=0.5, label='Group 2 (Variance = {:.2f})'.format(var2))
plt.title('Comparison of Two Groups')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

Q-27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results ?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

observed = np.array([30, 45, 25, 40, 60])

expected = np.array([40, 40, 40, 40, 40])

chi_square_stat = np.sum((observed - expected)**2 / expected)

df = len(observed) - 1


p_value = 1 - stats.chi2.cdf(chi_square_stat, df)

print(f"Chi-square Statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed frequencies do not match the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies match the expected frequencies.")

categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
bar_width = 0.35
index = np.arange(len(categories))

plt.figure(figsize=(8, 6))
plt.bar(index, observed, bar_width, label='Observed', alpha=0.7)
plt.bar(index + bar_width, expected, bar_width, label='Expected', alpha=0.7)
plt.xlabel('Categories')
plt.ylabel('Frequencies')
plt.title('Observed vs Expected Frequencies')
plt.xticks(index + bar_width / 2, categories)
plt.legend(loc='upper right')
plt.show()
```
